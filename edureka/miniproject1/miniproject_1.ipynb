```python
#task5
def plot_moving_average(stock_data, ticker, window_size=20):
    plt.figure(figsize=(12, 6))
    stock_data['Close'].plot(label=f'{ticker} Close Price', color='blue')
    stock_data['Close'].rolling(window=window_size).mean().plot(label=f'{ticker} Moving Average', color='red')
    plt.title(f'{ticker} Stock Price and Moving Average')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    plt.show()

plot_moving_average(AMZN, 'AMZN')
plot_moving_average(AAPL, 'AAPL')
plot_moving_average(GOOGL, 'GOOGL')
plot_moving_average(MSFT, 'MSFT')
```

```python
#task6
def calculate_daily_return_average(stock_data, ticker):
    daily_return = stock_data['Close'].pct_change()
    average_daily_return = daily_return.mean()
    print(f'Average Daily Return for {ticker}: {average_daily_return:.4%}')


calculate_daily_return_average(AMZN, 'AMZN')
calculate_daily_return_average(AAPL, 'AAPL')
calculate_daily_return_average(GOOGL, 'GOOGL')
calculate_daily_return_average(MSFT, 'MSFT')
```

```python
# task7
def add_trend_column(stock_data, ticker):
    stock_data['Daily_Return'] = stock_data['Close'].pct_change()
    stock_data['Trend'] = 'Up'
    stock_data.loc[stock_data['Daily_Return'] < 0, 'Trend'] = 'Down'

print(add_trend_column(AMZN, 'AMZN'))
print(add_trend_column(AAPL, 'AAPL'))
print(add_trend_column(GOOGL, 'GOOGL'))
print(add_trend_column(MSFT, 'MSFT'))
```

```python
# task8
def visualize_trend_pie(stock_data, ticker):
    trend_counts = stock_data['Trend'].value_counts()
    plt.figure(figsize=(8, 8))
    plt.pie(trend_counts, labels=trend_counts.index, autopct='%1.1f%%', startangle=90, colors=['green', 'red'])
    plt.title(f'{ticker} Trend Frequency')
    plt.show()


visualize_trend_pie(AMZN, 'AMZN')
visualize_trend_pie(AAPL, 'AAPL')
visualize_trend_pie(GOOGL, 'GOOGL')
visualize_trend_pie(MSFT, 'MSFT')
```
